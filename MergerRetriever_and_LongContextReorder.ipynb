# Now ingest the Data into the Chroma Database

```python
from langchain.vectorstores import Chroma
import chromadb
```

```python
import os
os.getcwd()
```

```python
CURRENT_DIR = os.path.dirname(os.path.abspath("."))
```

```python
CURRENT_DIR
```

```python
DB_DIR = os.path.join(CURRENT_DIR, "/content/db")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False,
)
```

```python
rls_vectorstore = Chroma.from_documents(text_rls,
                                       hf_bge_embeddings,
                                       client_settings=client_settings,
                                       collection_name="rls",
                                       collection_metadata={"hnsw":"cosine"},
                                       persist_directory="/store/rls")
```

```python
lim_vectorstore = Chroma.from_documents(text_lim,
                                       hf_bge_embeddings,
                                       client_settings=client_settings,
                                       collection_name="lim",
                                       collection_metadata={"hnsw":"cosine"},
                                       persist_directory="/store/lim")
```

 # Now Crearte a Retriever

```python
retriever_lim=lim_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
retriever_rls = rls_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

# Let's Merge both Retriever

# It is also called lord of retriever(LOTR)

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_rls, retriever_lim])
```

```python
for chunks in lotr.get_relevant_documents("whats lost in middle meaning?"):
    print(chunks.page_content)
```

```python
for chunks in lotr.get_relevant_documents("how optmisezers update and effect on leasrning rate?"):
    print(chunks.page_content)
```

# Now After understanding step by step it create a pipeline for LLM

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
)
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import ContextualCompressionRetriever
from langchain.document_transformers import LongContextReorder
```

```python
from re import search
filter = EmbeddingsRedundantFilter(embeddings=hf_bge_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr,search_kwargs={"k": 3, "include_metadata": True}
)
```

```python
docs = compression_retriever_reordered.get_relevant_documents("What is road less scedhuler meaning?")
print(len(docs))
print(docs[0].page_content)
```

```python
from langchain.llms import LlamaCpp
llms = LlamaCpp(streaming=True,
                   model_path="/content/zephyr-7b-beta.Q4_K_M.gguf",
                   max_tokens = 1500,
                   temperature=0.75,
                   top_p=1,
                   gpu_layers=0,
                   stream=True,
                   verbose=True,n_threads = int(os.cpu_count()/2),
                   n_ctx=4096)
```

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(
      llm=llms,
      chain_type="stuff",
      retriever = compression_retriever_reordered,
      return_source_documents = True
)
```

```python
query ="what is AlgoPerf challenge?"
results = qa(query)
print(results['result'])
#
print(results["source_documents"])
```
